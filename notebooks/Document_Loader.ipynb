```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

# 1. Initialize Loader
#loader = PyPDFLoader("./HSPM-14524750.pdf")
loader = PyPDFLoader("/Users/manikantaamara/Desktop/Antigravity/ai-engineering-visualized/ai-engineering-visualized/notebooks/HSPM-14524750.pdf")
# 2. Load the Document objects
docs = loader.load()

print(f"Loaded {len(docs)} pages.")
print(f"Metadata of page 1: {docs[0].metadata}")
# Output: {'source': 'annual_report.pdf', 'page': 0}

# 3. Split into manageable chunks
splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100)
chunks = splitter.split_documents(docs)

print(f"Split into {len(chunks)} chunks ready for embeddings.")
```
